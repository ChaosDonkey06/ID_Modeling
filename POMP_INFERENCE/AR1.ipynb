```python
import pandas as pd
import numpy as np

# stationary soluiton of the model

# x_t = sum(j=0)^inf phi_j * w_{t-j}

std_wt = 1
terms  = 10000
wts    = np.random.normal(loc=0, scale=std_wt, size=(terms))

phi    = 0.9
phi_j  = phi**np.arange(terms)


def ar1(xt, phi, wt):
    return phi * xt + wt

x0 = 2
x_all = np.full((terms, 1), np.nan)
x_all[0, :] = x0

for t in range(terms-1):
    x_all[t+1,:] = ar1(x_all[t,:], 0.9, wts[t])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(15.5, 7.2))
ax.plot(x_all[:100], lw=3, c="k")
```
